```python
#interactivity tester: golf scoring (lower is more intractive)
#interactivity between a and b = 1.0 if adjacent, -+ for each digit removed 
#interactivity between a and b  -1.0 if divisible without a remainder 


import numpy as np 
import pandas as pd 

start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))
count = start 

x=[]
while count < end:
    x.append(count)
    count+=1 


lenx = len(x)
total=[]
labels=[]

#print(x)
#print(lenx)

for number in x:
    for nextnumber in x: 
        if number == nextnumber:
            relationship = 0 
        else: 
            relationship = nextnumber - number
        if relationship < 0:
            relationship = relationship * -1 
        if nextnumber % number == 0:
            relationship=relationship-1
        elif nextnumber % number != 0: 
            relationship=relationship 
        total.append(relationship)
        label = nextnumber,'&',number,':'
        labels.append(label)
        
        
final = list(zip(labels,total))

df = pd.DataFrame(final, columns =['Number pair','Interactivity Score'])
print(df)

#for object in final:
#    print(object,'\n')
```
